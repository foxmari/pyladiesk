### 3. データの数値化して概要を知る

```python
#ラーメン店舗数が一番多いのはどの県？（ラーメン店舗の多い順に並び替えてみよう）
df.sort_values(by=['ramen_shop'], ascending=False)  #ascending=Trueにすると降順
```

```python
#要約統計量を算出しよう
df.describe()
```

```python
#新しい列を計算して作成する
df['female_%'] = df['female_k']/df['population_k']
```

```python
df.head()
```

```python
#度数集計をする
df['地方'].value_counts(dropna=False)
```

```python
#グルーピングする
grouped = df.groupby('地方')
```

```python
#グループ化した「地方」別に、要約統計量を算出してみる
grouped['ramen_shop'].describe()
```

### 4. matplotlibを使って可視化

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#地域別都道府県数の棒グラフ
sns.countplot('chihou', data=df)
```

```python
#ラーメン店数のヒストグラム
df.plot(y=['ramen_shop'], bins=20, alpha=0.5, figsize=(16,4), kind='hist')
```

```python
#平均値を確認してみる
df['ramen_shop'].mean()
```

```python
#中央値を確認してみる
df['ramen_shop'].median()
```

```python
#データのばらつきや歪みを確認するもう一つのグラフ：箱ひげ図
sns.boxplot('ramen_shop',data=df)
```

```python
#地方別に箱ひげ図を描いてみよう
sns.boxplot(x='chihou', y='ramen_shop',data=df)
```

```python
#人口とラーメン店舗数の関係を散布図で表してみよう
sns.jointplot('population_k','ramen_shop', data=df)
```

```python
#面積とラーメン店舗数の関係だとどうだろう？
sns.jointplot('area_km2','ramen_shop', data=df)
```
